```python
import numpy as np
import matplotlib.pyplot as plt
import math

#lat and longitude points in degrees
latitude = [50, 50, 50, 50, 50, 50, 60, 60, 60, 60, 60, 60, 70, 70, 70, 70, 70, 70, 80, 80, 80, 80, 80 ,80]
longitude = [-30, -20, -10, 10, 20, 30, -30, -20, -10, 10, 20, 30, -30, -20, -10, 10, 20, 30, -30, -20, -10, 10, 20, 30]

#degrees to radians
lon_arr = np.array(longitude)
lat_arr = np.array(latitude)
print(lon_arr)
print(lat_arr)
# Azimuthal equidistant projection parameters
R = 1  # Earth's radius in kilometers

#compute x and y coords in the a.e.p
lat_rad=(((math.pi)/180)* lat_arr)
lon_rad=(((math.pi)/180)* lon_arr)

rho=R * (((math.pi)/2)- lat_rad)
print(math.pi/2)
print(f'Rho: {rho}')
x = rho * np.sin(lon_rad)
y = - (rho) * np.cos(lon_rad)

print(f' Longitude{x}')
print(f' Latitude{y}')

# Scatter plot
plt.figure(figsize=(10, 7))
plt.scatter(x, y ,c='blue', marker='o')
plt.title('Azimuthal Equidistant Projection')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.axis('equal')

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given latitude and longitude points in degrees
longitude = [50, 50, 50, 50, 50, 60, 60, 60, 60, 60, 70, 70, 70, 70, 70]
latitude = [-30, -10, 10, 20, 30, -30, -10, 10, 20, 30, -30, -10, 10, 20, 30]

# Convert degrees to radians
lon_rad = np.radians(longitude)
lat_rad = np.radians(latitude)

# Azimuthal equidistant projection parameters
R = 6371  # Earth's radius in kilometers
lat0 = 0  # Latitude of the center of the projection in radians
lon0 = 0  # Longitude of the center of the projection in radians

# compute rho
cos_c = np.sin(lat0) * np.sin(lat_rad) + np.cos(lat0) * np.cos(lat_rad) * np.cos(lon_rad - lon0)
#cos_c = np.clip(cos_c, -1, 1)  # Clip values to avoid numerical issues
rho = R * np.arccos(cos_c)

#compute theta
theta = np.arctan2(np.sin(lon_rad - lon0) * np.cos(lat_rad), np.cos(lat0) * np.sin(lat_rad) - np.sin(lat0) * np.cos(lat_rad) * np.cos(lon_rad - lon0))


# Plotting the points in Cartesian coordinates using scatter plot
plt.figure(figsize=(10, 7))
plt.scatter(theta, rho, c='blue', marker='o')
plt.title('Azimuthal Equidistant Projection (Cartesian Scatter Plot)')
plt.xlabel('x (km)')
plt.ylabel('y (km)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```
